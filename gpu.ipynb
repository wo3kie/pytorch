```python
import torch
from torch import tensor
```

```python
assert(torch.cuda.is_available())

tensor = tensor([1, 2, 3])
assert(tensor.device==torch.device('cpu'))

tensor = tensor.to('cuda')
assert(tensor.device==torch.device('cuda:0'))

tensor = tensor.to('cpu')
assert(tensor.device==torch.device('cpu'))
```
