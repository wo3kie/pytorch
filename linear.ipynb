```python
from torch import tensor
import torch
```

```python
#
# torch.nn.linear
#
# aplies a linear transformation to the incoming data: y = x @ A.T + b
#
```

```python
torch.manual_seed(0)
linear = torch.nn.Linear(in_features=3, out_features=1)

assert(linear.weight.ndim == 2)
assert(linear.weight.shape == (1, 3))
assert(torch.allclose(linear.weight, tensor([[-0.0043,  0.3097, -0.4752]]), rtol=0.001, atol=0.001))

assert(linear.bias.ndim == 1) 
assert(linear.bias.shape == (1,))
assert(torch.allclose(linear.bias, tensor([-0.4249]), rtol=0.001, atol=0.001))


x = torch.tensor([1.0, 2.0, 3.0])
y = linear(x)
assert(y.ndim == 1)
assert(y.shape == (1,))
assert(torch.allclose(y, x @ linear.weight.T + linear.bias))
```

```python
torch.manual_seed(0)
linear = torch.nn.Linear(in_features=3, out_features=2)

assert(linear.weight.ndim == 2)
assert(linear.weight.shape == (2, 3))
assert(torch.allclose(linear.weight, tensor([[-0.0043,  0.3097, -0.4752],
                                             [-0.4249, -0.2224,  0.1548]]), rtol=0.001, atol=0.001))

assert(linear.bias.ndim == 1)
assert(linear.bias.shape == (2,))
assert(torch.allclose(linear.bias, tensor([-0.0114,  0.4578]), rtol=0.001, atol=0.001))


x = torch.tensor([1.0, 2.0, 3.0])
y = linear(x)
assert(y.ndim == 1)
assert(y.shape == (2,))
assert(torch.allclose(y, x @ linear.weight.T + linear.bias))
```
